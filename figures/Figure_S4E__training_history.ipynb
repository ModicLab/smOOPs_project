```python
import json
import matplotlib.pyplot as plt
import numpy as np
plt.rcParams["svg.fonttype"] = "none" 
```

```python
def plot_training_history(json_file):

    with open(json_file, 'r') as f:
        history = json.load(f)
    
    loss = history['loss']
    val_loss = history['val_loss']
    accuracy = history['accuracy']
    val_accuracy = history['val_accuracy']
    auc = history['auc']
    val_auc = history['val_auc']
    
    min_val_auc_epoch = np.argmax(val_auc)
    
    fig, axs = plt.subplots(3, 1, figsize=(10, 10))
    color_train = '#6294AC'
    color_val = '#D26E2F'
    color_line = "#D73C59"

    axs[0].plot(loss, label='Training Loss', color=color_train)
    axs[0].plot(val_loss, label='Validation Loss', color=color_val)
    axs[0].axvline(min_val_auc_epoch, linestyle='--', color=color_line, label=f'Min Val AUC Epoch: {min_val_auc_epoch+1}')
    axs[0].set_ylabel('Loss', fontsize=14)
    axs[0].legend()
    axs[0].set_xlim(0, len(loss))
    
    axs[1].plot(accuracy, label='Training Accuracy', color=color_train)
    axs[1].plot(val_accuracy, label='Validation Accuracy', color=color_val)
    axs[1].axvline(min_val_auc_epoch, linestyle='--', color=color_line, label=f'Min Val AUC Epoch: {min_val_auc_epoch+1}')
    axs[1].set_ylabel('Accuracy', fontsize=14)
    axs[1].legend()
    axs[1].set_xlim(0, len(loss))
    
    axs[2].plot(auc, label='Training AUC', color=color_train)
    axs[2].plot(val_auc, label='Validation AUC', color=color_val)
    axs[2].axvline(min_val_auc_epoch, linestyle='--', color=color_line, label=f'Min Val AUC Epoch: {min_val_auc_epoch+1}')
    axs[2].set_xlabel('Epoch', fontsize=14)
    axs[2].set_ylabel('AUC', fontsize=14)
    axs[2].legend()
    axs[2].set_xlim(0, len(loss))
    
    plt.tight_layout()
    plt.show()

plot_training_history('Model_Training/Models/Final_Training_Run/training_run_19092024_085523219/postar3/saved_trained_model/training_history.json')
plot_training_history('Model_Training/Models/Final_Training_Run/training_run_12092024_163543175/sequence:/saved_trained_model/training_history.json')
```

```python
import pandas as pd
import subprocess
```

```python
naive_transcripts = pd.read_csv('Data/machine_learning_input_prep/all_transcripts_with_fasta_clip_m6a_paris_intra_paris_inter_postar3_rnafold_collapsed.bed', sep="\t")
naive_transcripts["sequence_length"] = naive_transcripts["sequence"].str.len()
```

```python
def create_ones_string(sequence: str) -> str:
    string = '1' * len(sequence)
    return ','.join(string)

naive_transcripts['length_control'] = naive_transcripts['sequence'].apply(create_ones_string)

naive_transcripts
```

```python
naive_transcripts.to_csv('Data/machine_learning_input_prep/naive_transcripts_with_fasta_clip_m6a_paris_intra_paris_inter_postar3_rnafold_length_control_collapsed.bed', sep="\t", index=False)
```

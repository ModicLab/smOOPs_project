```python
naive_transcripts = pd.read_csv('Data/machine_learning_input_prep/all_transcripts_with_fasta_clip_m6a_paris_intra_paris_inter_postar3_collapsed.bed', sep="\t")
naive_transcripts
```

```python
import pandas as pd
import subprocess
from multiprocessing import Pool

def_new = pd.DataFrame()

def predict_structure(dna_sequence):
    print(len(dna_sequence))
    rna_sequence = dna_sequence.replace('T', 'U')
    result = subprocess.run(['RNAfold'], input=rna_sequence, encoding='ascii', stdout=subprocess.PIPE, stderr=subprocess.PIPE)
    
    output = result.stdout.split('\n')
    structure = output[1].split(' ')[0]  # Structure part of the output
    energy = output[1].split(' ')[1]  # Energy part of the output
    print(energy)
    return structure, energy, dna_sequence


df = naive_transcripts.sort_values('sequence_length', ascending=True)

with Pool(40) as pool:
    results = pool.map(predict_structure, df['sequence'])
    pool.close()
    pool.join()
    
structures, energies, dna_sequence = zip(*results)

def_new['structure'] = structures
def_new['energy'] = energies
def_new["sequence_new"] = dna_sequence
```

```python
merged_df = pd.merge(df, def_new, left_on='sequence', right_on='sequence_new')
merged_df = merged_df.drop(columns=['sequence_new'])
```

```python
def average_stem_length(dot_bracket):
    stack = []
    stem_lengths = []

    for char in dot_bracket:
        if char == '(':
            stack.append(char)
        elif char == ')':
            if stack:
                stem_lengths.append(len(stack))
                stack.pop()

    if stem_lengths:
        return sum(stem_lengths) / len(stem_lengths)
    else:
        return 0

merged_df['Average Stem Length'] = merged_df['structure'].apply(average_stem_length)
merged_df["Energy Formatted"] = merged_df["energy"].str.replace("(", "").str.replace(")", "").astype(float)
merged_df["Energy Density"] = merged_df["Energy Formatted"] / merged_df["sequence_length"]
merged_df["Number of Structured Nucleotides"] = merged_df["structure"].str.count("\(")
merged_df["Number of Structured Nucleotides Density"] = merged_df["Number of Structured Nucleotides"] / merged_df["sequence_length"]
```

```python
merged_df.to_csv('Data/machine_learning_input_prep/naive_transcripts_with_fasta_clip_m6a_paris_intra_paris_inter_postar3_rnafold-raw.bed', sep="\t", index=False)
```

```python
def encode_structure(structure: str) -> str:
    encoded = structure.replace('.', '0').replace('(', '1').replace(')', '1')
    return ','.join(encoded)

naive_transcripts['rnafold'] = naive_transcripts['structure'].apply(encode_structure)

naive_transcripts.drop(columns=['structure', 'energy','sequence_length', 'Average Stem Length'	,'Energy Formatted',	'Energy Density',	'Number of Structured Nucleotides',	'Number of Structured Nucleotides Density'], inplace=True)
```
